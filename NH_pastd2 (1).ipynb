## 1. 일부매수, 일부 매도 기록 빼기

```python
train_1=train.copy()
hist_1=hist.copy()
train_1['past_d']=0
```

```python
train_1.rename(columns={'계좌ID':'act_id', '종목코드': 'iem_cd', '매수일자':'byn_dt'}, inplace=True)
```

```python
train_1.head()
```

```python
for i in range(30000,50000):
  a=train_1['act_id'][i]
  b=train_1['iem_cd'][i]
  selected_train=train_1[(train_1['act_id']==a) & (train_1['iem_cd']==b)]
  selected_hist=hist_1[(hist_1['act_id']==a) & (hist_1['iem_cd']==b)]
             
  # hist기록이 곧 train 기록인 경우, past_d=hold_d
  if (selected_hist.shape[0] < 4):
    pass
  # train 매수 시점 이전에 전량매도가 없는 경우
  else:
    zero_hist=selected_hist[selected_hist['bnc_qty']==0]
    if zero_hist.shape[0] < 2:
      pass
    # train 매수 시점이전에 전량매도가 있는 경우
    elif zero_hist.shape[0] > 1:
      zero=[]
      first=[]
      first.append(selected_hist['bse_dt'].iloc[0])

      for a in range(len(selected_hist)-1):
        if selected_hist['bnc_qty'].iloc[a]==0:
          zero.append(selected_hist['bse_dt'].iloc[a])
          first.append(selected_hist['bse_dt'].iloc[a+1])
      
      del first[-1]

      sum_date=[]
      for b in range(len(zero)):
        last_hist=pd.to_datetime(str(zero[b])).strftime("%Y-%m-%d")
        first_train=pd.to_datetime(str(first[b])).strftime("%Y-%m-%d")
        sum_date.append(np.busday_count(first_train, last_hist))
      
      train_1['past_d'][i]=np.round(sum(sum_date)/(zero_hist.shape[0]-1),2)
      
      zero=[]
      hist=[]
      sum_date=[]
```

```python
train_1.to_csv("train30000.csv")
```
